```python
import re
import zipfile

class ZipBook():
    
    def __init__(self, zipfilename):
        """Open book from ZIP based on a path."""
        self.filename = zipfilename
        self.getContents()
        self.getLines()
        self.getStartEndLocations()
        self.headingLines = []
    
    def getContents(self):
        """Reads the book into memory."""

        with zipfile.ZipFile(self.filename, 'r') as book_zip:
            try:
                textname = book_zip.namelist()[0]
                # We might want to change this to open the one txt file in the zip
                with book_zip.open(textname) as txtfile:
                    text = txtfile.read().decode()
            except IndexError:
                text = ""
            
        self.contents = text
    
    def getLines(self):
        """Strip into lines."""
        self.lines = self.contents.split('\n')
        
    @property
    def bookLines(self):
        """Return only lines of bookText"""
        return self.lines[self.startLocation+1:self.endLocation]
        
    def bookText(self):
        """Return cleaned text."""
        return "\n".join(self.bookLines)

    def getStartEndLocations(self):
        """
        Tries to find where the book starts and ends.
        """
        starts = ["Start of the Project Gutenberg EBook",
                "Start of Project Gutenberg's",
                "\*\*\*START OF THE PROJECT GUTENBERG EBOOK",
                "\*\*\* START OF THIS PROJECT GUTENBERG EBOOK"]
        ends = ["End of the Project Gutenberg EBook",
                "End of Project Gutenberg's",
                "\*\*\*END OF THE PROJECT GUTENBERG EBOOK",
                "\*\*\* END OF THIS PROJECT GUTENBERG EBOOK"]
        joined_starts = '|'.join(starts)
        joined_ends = '|'.join(ends)
        pat_start = re.compile(joined_starts, re.IGNORECASE)
        pat_end = re.compile(joined_ends, re.IGNORECASE)
        self.startLocation = 0
        self.endLocation = len(self.lines)
        for line in self.lines:
            if pat_start.match(line) is not None:
                self.startLocation = self.lines.index(line)
            if pat_end.match(line) is not None:
                self.endLocation = self.lines.index(line)
                break
                
    def getHeadingLines(self):
        """Apply regex to detect heading lines
        
        Stored in headingLines as (start, stop)
        """
        self.headingLines = []
        heading = "^[A-Z_]{1,50}[^a-z,]+\s{0,3}$"
        pat = re.compile(heading)
        for line in self.bookLines:
            if pat.match(line):
                # Start and stop are on same line initially
                self.headingLines.append(
                    self.lines.index(line), self.lines.index(line)
                )
    
    def getHeadings(self):
        """Return line text of headings"""
        if not self.headingLines:
            self.getHeadingLines()
        return [
            "\n".join(self.lines[s:e]) for s, e in self.headingLines
        ]
    
    def joinHeadingLines(self):
        """Combine lines of a common heading."""
        revisedHeadings = []
        runningStart = self.headingLines[0]
        for i, s, e in enumerate(self.headingLines[:-1]):
            if s == e:
                j = s
                blankline = True
                while not self.lines[j].strip():
                    j++
                if j == self.headingLines[i+1]:
                    end = e
                else:
                    revisedHeadings.append((runningStart, end))
                    runningStart
    
    def getChapterText(self):
        chapters = []
        lastHeading = len(self.headingLocations) - 1
        for i, headingLocation in enumerate(self.headingLocations):
            # We need to look for pairs of heading that
            # contain the chapter number and title
            j = i
            blankline = True
            while blankline:
                j++
                if not self.lines[j].strip() or j = nextheading:
                    blankline = False
            if j != i:
                # Combine headings
                
            if i is not lastHeading:
                nextHeadingLocation = self.headingLocations[i+1]
                chapters.append(self.lines[headingLocation+1:nextHeadingLocation])
        return chapters
    
    def ignoreTOC(self):
        """
        Filters headings out that are too close together,
        since they probably belong to a table of contents.
        """
        pairs = zip(self.headingLocations, self.headingLocations[1:])
        toBeDeleted = []
        for pair in pairs:
            delta = pair[1] - pair[0]
            if delta < 4:
                if pair[0] not in toBeDeleted:
                    toBeDeleted.append(pair[0])
                if pair[1] not in toBeDeleted:
                    toBeDeleted.append(pair[1])
        logging.debug('TOC locations to be deleted: %s' % toBeDeleted)
        for badLoc in toBeDeleted:
            index = self.headingLocations.index(badLoc)
            del self.headingLocations[index]
            
        
```
